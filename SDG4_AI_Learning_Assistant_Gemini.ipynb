```python
import google.generativeai as genai

for m in genai.list_models():
    print(m.name, "-> supports generateContent:", "generateContent" in m.supported_generation_methods)
```

```python
def learning_assistant(topic, level):
    prompt = f"""
    You are an AI-powered personalized learning assistant.

    Topic: {topic}
    Learner Level: {level}

    Adapt your explanation based on the learner level:
    - School: very simple language, basic concepts
    - College: academic explanation with examples
    - Beginner: step-by-step basics
    - Intermediate: moderate depth with applications
    - Advanced: in-depth theoretical explanation
    - Expert: concise, technical, insight-driven explanation

    Tasks:
    1. Explain the topic clearly according to the learner level.
    2. Give one relevant real-world or practical example.
    3. Generate 5 multiple-choice questions (MCQs) appropriate to the level.
    4. Provide the correct answers at the end.

    Keep the content educational and well-structured.
    """

    model = genai.GenerativeModel("models/gemini-flash-lite-latest")
    response = model.generate_content(prompt)
    return response.text
```

```python
topic = input("Enter the topic you want to learn: ")
level = input(
    "Enter student level (School / College / Beginner / Intermediate / Advanced / Expert): "
)

output = learning_assistant(topic, level)

print("\n========== AI LEARNING ASSISTANT OUTPUT ==========\n")
print(output)
```
